```python
import pandas as pd
a = pd.Series([1, 2, 3])
b = pd.Series([4, 5, 6])

# Вычисляем Евклидово расстояние вручную
distance = np.sqrt(((a - b) ** 2).sum())

print("Евклидово расстояние:", distance)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Пример данных
df = pd.read_csv('https://raw.githubusercontent.com/akmand/datasets/master/iris.csv')
minmax_scaler = MinMaxScaler()
df["sepal_length_scaled"] = minmax_scaler.fit_transform(df[["sepal_length_cm"]])

z_scaler = StandardScaler()
df["sepal_width_scaled"] = z_scaler.fit_transform(df[["sepal_width_cm"]])

# Вывод результата
print(df)
```
